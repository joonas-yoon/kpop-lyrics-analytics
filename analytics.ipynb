```python
import pandas as pd
import os
import sys
```

```python
files = os.listdir('datasets')

alls = {
    'id': [],
    'year': [],
    'title': [],
    'singer': [],
    'lyric': []
}

for f in sorted(files):
    df = pd.read_csv(os.path.join('datasets', f), delimiter='|', encoding='utf-8')
    year = int(f.replace('MELON_TOP100_', '').replace('.csv', ''))
    if not 'id' in df.columns:
        df = df.transpose()
        df.columns = list(df.iloc[0])
        df.drop(index='Unnamed: 0', inplace=True)

    try:
        alls['id'] += list(df['id'])
        if not 'year' in df.columns:
            years = [year] * len(df)
        else:
            years = list(df['year'])
        alls['year'] += years
        alls['title'] += list(df['title'])
        alls['singer'] += list(df['singer'])
        alls['lyric'] += list(df['lyric'])
    except Exception as e:
        print(e, files=sys.stderr)
        print(df)

df = pd.DataFrame.from_dict(alls)
df
```

```python
raw_df = df.copy()
df = raw_df.sort_values(by=['year', 'id']).reset_index(drop=True)
df['id'] = df['id'].astype(int)
df['year'] = df['year'].astype(int)
df
```

```python
import re

def remove_special_chars(text):
    return re.sub('[?!<>@#$\~\\/\.&*\-\+=_,:;\{\}\[\]]', '', text)

def cleanize_title(text):
    if not text or type(text) != str:
        return ''
    text = remove_special_chars(text)
    text = re.sub('[\(\)]', '', text)
    return text

def cleanize_lyrics(text):
    if not text or type(text) != str:
        return ''
    text = remove_special_chars(text)
    if text.startswith("[가사 준비중]"):
        return ''
    if text.endswith("펼치기"):
        text = text[:-len("펼치기")]
    return text

df['lyric'] = df['lyric'].apply(cleanize_lyrics)
df['title'] = df['title'].apply(cleanize_title)
df['singer'] = df['singer'].apply(cleanize_title)
df
```

```python
h = Hannanum()
okt = Okt()

def get_nouns(text):
    return h.nouns(text)

def get_okt(text):
    # 형태소를 원형으로 처리 + 원형으로 교체
    return okt.pos(text.replace('\n', ' '), norm=True, stem=True)

def filter_unicode(text, unicode_range):
    start, end = unicode_range
    return ''.join(filter(lambda c: start <= c <= end, text))

def to_single_space(text):
    return re.sub('\s{2,}', ' ', text).strip()

def filter_language_raw(text, lang):
    if lang == 'kr':
        return re.sub('[^\uac00-\ud7a3\s]+', '', text)
    elif lang == 'cn':
        return re.sub('[^\u4e00-\u9fa5\s]+', '', text)
    elif lang == 'jp':
        return re.sub('[^\u0800-\u4e00\s]+', '', text)
    return re.sub('[^a-zA-Z\s]+', '', text)

def filter_language(text, lang = 'kr'):
    return to_single_space(filter_language_raw(text, lang))
```

```python
from tqdm.notebook import tqdm

counter = Counter()
lyrics = df['lyric']
bar = tqdm(total=len(lyrics))
for lyric in lyrics:
    l_ko = filter_language(str(lyric), 'kr')
    sentence = [f"{word}_{ty}_kr" for word, ty in get_okt(l_ko)]
    c = Counter(sentence)
    counter += c
    del c
    bar.update(1)

print(counter)
```

```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import word_tokenize

nltk.download('popular')

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
```

```python
for l in tqdm(lyrics):
    l = '' if type(l) != str else l
    sentence = filter_language(l, 'en')
    if len(sentence) < 1:
        continue
    words = sentence.split()
    c = Counter([f"{lemmatizer.lemmatize(word, 'v')}_Eng_en" for word in words])
    counter += c
    del c
```

```python
count_pairs = []

for key, count in counter.items():
    word, kind, lang = key.split('_')
    count_pairs.append((word, kind, count, lang))

count_df = pd.DataFrame(count_pairs, columns=['word', 'type', 'count', 'language'])
count_df = count_df.sort_values(by=['count', 'type'], ascending=[False, True])
count_df
```

```python
def get_words_counts(df) -> dict:
    d = df[['word', 'count']]
    d.index = d['word']
    d = d.drop(['word'], axis=1)
    return d.to_dict()['count']
```

```python
from wordcloud import WordCloud

fonts_dir = "notosanskr"
fonts = list(filter(lambda f: f.endswith(".otf"), os.listdir(fonts_dir)))
fonts = list(map(lambda f: os.path.join(fonts_dir, f), fonts))
```

```python
is_kr = count_df['language'] == 'kr'
is_en = count_df['language'] == 'en'
is_noun = count_df['type'] == 'Noun'
is_verb = count_df['type'] == 'Verb'
is_adj = count_df['type'] == 'Adjective'
```

```python
def save_word_cloud(filename: str, data: dict, max_font_size=128, width=512, height=512):
    wc = WordCloud(font_path=fonts[0], background_color="white", max_font_size=max_font_size, width=width, height=height)
    cloud = wc.generate_from_frequencies(data)
    cloud.to_file(filename)
    return cloud
```
